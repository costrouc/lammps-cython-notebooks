```python
m = structure.lattice.matrix
```

```python
a = lammps.core.lengths_angles_to_matrix(*structure.lattice.abc, *(np.array(structure.lattice.angles) * (math.pi/180.0)))
a
```

```python
b = np.linalg.solve(a, m)
```
